```python

bins = np.linspace(min(df["Year"]), max(df["Year"]), 4)
group_names = ['Old', 'Medium', 'Modern']
df['year-binned'] = pd.cut(df['Year'], bins, labels=group_names, include_lowest=True )
df[['Year','year-binned']].head(20)
```

```python
# %matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["year-binned"].count())
```
